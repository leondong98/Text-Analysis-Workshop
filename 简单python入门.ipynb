### **Python 主要的数据容器**

Python 提供了几种内建的数据容器, 帮助你高效地存储和组织数据。对于初学者来说, 最常见的四种是: 列表(list)、字典(dictionary)、元组(tuple)和集合(set)。


- **列表** (`[]`): **有序的、可变的元素集合。** 列表可以包含不同类型的元素，创建后可以修改、添加或删除其中的内容。
示例: `my_list = [1, 'apple', True]`

- **字典** (`{}`): **无序的键-值对集合。** 每个键(key)必须唯一，字典用于通过键快速查找或存储数据.
示例: `my_dict = {'name': 'Trump', 'age': 78}`

- **元组** (`()`): **有序的、不可变的元素集合。** 与列表类似，元组可以存储多种类型的值，但创建后内容不可更改。
示例: `my_tuple = (10, 'banana', False)`

- **集合** (`{}`): **无序的、唯一元素的集合。** 集合自动去除重复值，适用于进行成员测试或消除冗余数据的场景。
示例: `my_set = {1, 1, 2, 3} → {1, 2, 3}`

这些数据容器各有优势，掌握它们的使用场景是编写简洁高效 Python 代码的关键一步。

### **Python 索引机制 (Indexing)**

在 Python 中, 索引是访问数据容器中单个元素的方式, 适用于列表(list)、元组(tuple)、字符串(string)和字典(dictionary)等数据结构。

Python 使用从 0 开始的索引(zero-based indexing), 也就是说:

  -	第一个元素的位置是 0, 第二个是 1, 依此类推。

  -	同时，你也可以使用**负索引(negative indexing)**从序列的末尾开始计数，其中 -1 表示最后一个元素。

列表、元组、字符串 等顺序类型(sequences)使用位置索引(positional indexing), 索引必须是整数。

  - 如果访问一个不存在的索引位置，将会引发索引错误 (IndexError)。

与此不同, 字典(dictionary)使用的是键索引(key-based indexing), 而不是位置索引。
你通过键来访问对应的值，而不是通过数字位置。

&nbsp;


| 数据类型       | 索引类型        | 是否有顺序 | 支持负索引 | 索引是否为整数 | 示例                          |
|----------------|-----------------|------------|------------|----------------|-------------------------------|
| 列表 (List)    | 位置索引        | 是         | 是         | 是             | `my_list[0]`, `my_list[-1]`   |
| 元组 (Tuple)   | 位置索引        | 是         | 是         | 是             | `my_tuple[1]`, `my_tuple[-2]` |
| 字符串 (String)| 位置索引        | 是         | 是         | 是             | `'hello'[0]`, `'hello'[-1]`   |
| 字典 (Dict)    | 键索引（key）   | 否         | 否         | 否（使用键）   | `my_dict['name']`             |
| 集合 (Set)     | 不支持索引      | 否         | 否         | 否             | ❌ 不支持直接索引访问         |

````python
#@title Indexing Example

my_list[0] # 第一个元素
my_list[1] # 第二个元素
my_list[-1] # 最后一个元素

ex_dict['a'] # 索引键 “a” 对应的值

my_list = ['a', 'b', 'c']
print(my_list[0])   # 输出: 'a'
print(my_list[-1])  # 输出: 'c'

my_dict = {'name': 'Trump', 'age': 78}
print(my_dict['name'])  # 输出: 'Trump'
````

### **Python 切片 (Slicing)**

**切片是从序列中访问一部分元素的方式** 适用于列表(list)、元组(tuple)和字符串(string)等序列类型。

基本语法是: `sequence[start:stop]`

 - 返回从 start 索引开始、到 stop 索引之前(不包括 stop)的一个新子序列。
 - 注意：切片中的 stop 是“左闭右开”的，即包括起始，不包括结束。

你还可以加入步长参数, 语法是: `sequence[start:stop:step]`

- step 表示每次移动多少个索引单位，默认是 1。
- step 可以为负值，这样就可以实现反向切片。

````python
#@title Slicing Example
my_list = [0, 1, 2, 3, 4, 5]

print(my_list[1:4])      # 输出 [1, 2, 3]
print(my_list[:3])       # 输出 [0, 1, 2]
print(my_list[::2])      # 输出 [0, 2, 4]
print(my_list[::-1])     # 输出 [5, 4, 3, 2, 1, 0]，反转列表
````

### **For 与 While 循环 (For and While Loops)**

循环语句让你可以重复执行一段代码块, 是实现自动化、遍历数据和控制流程的重要工具。Python 中最常用的两种循环结构是: for 循环 和 while 循环。

&nbsp;

**🔁 for 循环**

当你需要遍历 (iterate)一个序列（如列表、字符串或数字范围）时使用 for 循环。

```python
for item in [1, 2, 3]:
    print(item)
```
output:
```python
1
2
3  
```

&nbsp;

**🔁 while 循环**

当你不确定需要循环多少次，仅知道某个条件为真时要循环时，使用 while 循环。

```python
x = 0
while x < 3:
    print(x)
    x += 1
```
output:
```python
0
1
2
```

&nbsp;

**🔑 Python 中的循环关键字(保留字):**

`for`, `while`, `in`, `break`等，是 Python 语言的保留关键字(reserved keywords), 具有特殊含义, 不能用作变量名:

&nbsp;

| 特性                | `for` 循环                                | `while` 循环                                 |
|---------------------|--------------------------------------------|----------------------------------------------|
| 用途                | 遍历序列（list、tuple、string、range 等）   | 在某个条件为真时反复执行                      |
| 是否需要初始化变量  | 否（自动从序列中取值）                    | 是（通常需要初始化计数器变量）               |
| 适用场景            | 已知迭代次数、处理集合数据                  | 条件控制循环，如等待用户输入、监测状态等     |
| 是否需指定终止条件  | 否（由序列长度决定）                       | 是（显式指定条件）                           |
| 可用关键词          | `for`, `in`, `break`, `continue`           | `while`, `break`, `continue`                |
| 可否无限循环        | 否（除非使用 `while True` 的等价写法）       | 可以 (`while True:` 创建死循环)              |


### **条件语句(Conditional Statements)**

条件语句允许程序**根据条件是否成立来决定执行哪一段代码**。这让你的程序能够进行“判断”和“分支”。

&nbsp;

🔒 基本结构说明：

-	`if`判断一个条件是否为真,若为真，则执行对应代码块。
-	`elif` (“else if”的缩写) 如果前面的 if 条件为假，尝试判断另一个条件。
- `else` 如果以上所有条件都为假,则执行这一部分代码。

&nbsp;

| 关键词     | 中文释义         | 用途说明                                 | 示例                           |
|------------|------------------|------------------------------------------|--------------------------------|
| `if`       | 如果             | 判断条件是否为真，执行代码块             | `if x > 0:`                    |
| `elif`     | 否则如果         | 上一个条件为假时，检查另一个条件         | `elif x == 0:`                 |
| `else`     | 否则             | 所有条件都为假时执行                     | `else:`                        |


````python
#@title Example Conditional Statement
x = 10

if x > 15:
    print("x is greater than 15")
elif x > 5:
    print("x is greater than 5 but not more than 15")
else:
    print("x is 5 or less")
````

**🌟小拓展：嵌套条件语句 (Nested if)与三元表达式 (One-liner if)**

**嵌套条件语句 (Nested if)**: 是指在一个 if 或 else 的代码块中，再写一个 if 判断，用于处理更复杂的决策逻辑。

```python

if x > 0:
    if x < 10:
        print("x 是一个小的正数")
    else:
        print("x 是一个较大的正数")
else:
    print("x 不是正数")
```
output:
```python

```

**三元表达式 (One-liner if)**: 也叫条件表达式 (Ternary Expression), 是一种简洁写法，适合在一行中进行简单判断和赋值。

```python

```


### **函数 (Function)**

````python
#@title Example Function

students = {
    "John": {"age": 20,
            "has_id": False},
    "Cook": {"age": 17,
            "has_id": True},
    "Ternus": {"age": 31,
               "has_id": True}
}

# IMPORTANT: INDENTATION!!!
def check_id(students: dict):
    for student in students.keys():
        age = students[student].get("age")
        id = students[student].get("has_id")

        if age >= 18 and id:
            print(f"{student}: access granted")
        else:
            print(f"{student}: no ")

check_id(students)
````

### **类的继承 (Class Inheritance)**

**继承（Inherence）**是面向对象编程中的核心概念之一，允许一个类（称为**子类**或**派生类**）继承另一个类（称为**父类**或**基类**）的属性和方法。

通过继承，子类可以重用父类已有的功能，同时添加或修改其行为，避免重复编写代码。

在 Python 中，通过在类定义的括号中传入父类名称来创建一个子类。如果子类需要自定义初始化过程，可以使用 `super()` 函数调用父类的 `__init__ `方法，然后添加自己的属性或行为。

继承有助于保持代码的组织性，减少重复，并实现灵活、可扩展的程序设计。


````python
#@title Example Inheretance

class GraduateStudent(Student):
    def __init__(self, name, subject, grade, has_thesis):
        # Use super() to call the parent class's __init__ method
        super().__init__(name, subject, grade)
        self.has_thesis = has_thesis

    def is_doctor(self):
        # Graduate students need at least 70 to pass
        return self.grade >= 70 and self.has_thesis

anna = Student("Anna", "Python", 67)
bob = GraduateStudent("Bob", "AI Ethics", 72, True)
chris = GraduateStudent("Chris", "AI Ethics", 68, True)

print(anna.has_passed())   # True (Student needs 51)
print(bob.has_passed())    # True (Student needs 51)
print(chris.has_passed())  # True (Student needs 51)

print(bob.is_doctor())    # True (Graduate needs 70 + thesis)
print(chris.is_doctor())  # False (68 is not enough)

````

## **NumPy**
numpy.png.webp

在使用大型数据集或进行数值计算时，**性能至关重要**。Python 中的标准数据容器（如列表）虽然灵活且易于使用，但在进行大量数学运算时会变得低效。

这时就要引入 **NumPy —— 一个强大的库，它引入了一种高性能的数据容器 ndarray（n维数组）**。
与 Python 的列表不同，ndarray 是专为数值计算设计的。它提供了更快的计算速度、更低的内存占用，并具备便捷的语法来进行数组操作、线性代数和统计计算。

数据科学和机器学习领域中的大多数主流库 —— 包括 Pandas、Matplotlib 和 Scikit-learn —— 都是建立在 NumPy 之上的。因此，学习 NumPy 是进入更高级工具和工作流的基础步骤。


### **NumPy好在哪里？**

Python 是动态类型语言（dynamically typed），意味着你不需要声明变量的类型 —— Python 会自动推断。虽然这让 Python 非常易读且灵活，但也带来了额外的开销（Overhead）：列表中的每个元素都是一个完整的 Python 对象，携带着如类型、内存引用等元数据。

对于小型任务，这种开销可以忽略不计。但在进行大规模数值运算时，这种开销会成为性能瓶颈 —— 导致运算更慢、内存占用更高。

> NumPy 通过其 ndarray 解决了这个问题，它是静态类型的数组容器 —— 所有元素都以相同的数据类型（如 float64）连续存储在内存中。这意味着：

- 不需要多余的类型检查；
- 每个元素不再单独存储类型元数据（type metadata）；
- 访问和运算速度更快（得益于底层的 C 语言实现）。


简而言之：**NumPy 用牺牲一部分灵活性换来了巨大的速度与效率提升**，使其成为 Python 中进行高性能数值计算的首选工具。

### **数组的索引与切片（Indexing and Slicing Arrays）**

#### **一维数组（1D Arrays）**
要从数组中选择某个元素，使用方括号并通过索引指定所需元素的位置（从 0 开始计数）

一维数组的索引方式与 Python 列表类似:

```
array[index]
```

#### **二维数组（2D Arrays）**

对于多维数组，要获取某个元素，需要为数组的每一维提供一个索引。

对于二维数组（matrix），使用一对索引 `(row, column)` 来访问：

```
array[row, column]
```

#### **三维数组及更高维（3D Arrays 及以上）**
对于维度更高的数组，可以将“列索引”视为最后一维，“行索引”为倒数第二维。

```
array[:, :, row, column] # 示例：用于一个 4D 数组
```

````python
#@title Example Array Indexing

arr_2d = np.arange(27).reshape((9,3))  #创建一个数字从0到27，9行三列的矩阵
arr_2d

arr_2d[0, 0] # get first row first column value
arr_2d[0, -1] # last column of first row
arr_2d[1, ] # all of second row

arr_3d = np.arange(27).reshape((3,3,3))
arr_3d

arr_3d[1] # get second element of first dimension (matrix)
arr_3d[1, 1] # get second matrix, second row
arr_3d[:, -1, -1] # get second matrix, second row, third column value
````

### **数组的切片（Slicing Arrays）**
要获取多个元素或子数组，我们仍然使用方括号（square brakets）和切片语法：


```
array[start:stop:step]
```

- `start` 起始索引（默认值为 0）
- `top` 终止索引（默认是该维度的长度，不包含该索引）
- `step` 步长（默认是 1）

如果将 `step` 设置为负值，`start` 和 `stop` 的默认值会被交换，这是一种**反转数组**的常见技巧。


> NumPy 的切片操作会创建数组的一个视图（view），而不是副本（copy）。这意味着对切片后的数组进行更改，会影响原数组中的对应部分。


| 操作            | 含义                         | 示例             | 说明/备注                     |
|-----------------|------------------------------|------------------|------------------------------|
| `array[i, j]`   | 访问第 i 行第 j 列的元素       | `a[1, 2]`        | 用于二维或更高维数组索引     |
| `::2`           | 步长为 2，隔一个取一个元素     | `a[::2]`         | 用于切片，返回偶数索引的元素 |
| `[::-1]`        | 倒序                           | `a[::-1]`        | 快速反转数组顺序             |
| `.copy()`       | 显式复制数组                   | `b = a.copy()`   | 防止修改视图影响原数组       |
| `.reshape()`    | 改变数组形状，不改变数据内容   | `a.reshape(2, 3)`| 新形状必须与元素数量匹配     |

````python
#@title Example Slicing Arrays

arr_2d = np.arange(27).reshape((3, 9))     # 创建一个从 0 到 26 的一维数组，重塑成 3 行 9 列的二维数组
arr_2d

arr_2d[1, :4] # 第 2 行（索引为 1），前 4 个元素
arr_2d[1, ::2] # 第 2 行，每隔一个取一个（步长为 2）
arr_2d[::2, ::2] # 每隔一行 & 每隔一列取一个，起点都是第 0 行/列

arr_2d[::-1, ] # 所有行反转（上下颠倒）

# view vs copy
second_row = arr_2d[1]  # 这并不是复制了第 2 行，而是创建了 一个对原数组的“视图”（view）
second_row

second_row[:2] = 999 # 注意！把视图中的前两个值改成了 999，原数组也会被修改！
second_row
arr_2d

#⚠️注意⚠️ NumPy 默认用“视图”而不是“复制”，这可以节省内存，但如果你要避免影响原数据，还是需使用 .copy()


# 创建三维数组
arr_3d = np.arange(27).reshape((3,3,3)) #创建一个形状为 (3,3,3) 的三维数组（3 个 3x3 的小矩阵）
arr_3d

arr_3d[1, 1, ::2] # 取第 2 个“矩阵”（索引为 1）它的第 2 行（索引为 1），这一行中，每隔一列取一个元素（步长为 2）
````

### **广播机制（Broadcasting）**

假设你有两个形状不同的数组，但你想对它们执行某种操作（例如相加）。通常你需要将它们“变为相同形状”才能进行逐元素运算。


> **广播（Broadcasting）** 是 NumPy 用来扩展或复制较小数组形状的机制，从而使它们在逐元素操作（element-wise operations）时形状相容。



**✅ 两个维度在以下情况下被认为是兼容的：**
1. **两个维度相等，或**
2. **其中一个维度为 1**


**🔄 广播的工作方式：从右向左比较数组的形状维度** （即从最后一个维度开始，逐步向前比较）

**📢 广播的步骤详解：**

1. **对齐维度（Align Dims）**: 如果两个数组的维度数量不同，NumPy 会在较小数组的形状前补上 1，直到两者维度数一致。

```
a.shape = (3, 4, 5)
b.shape = (4, 5) → prepend 1 to make (1, 4, 5)

```

2. **检查兼容性（Check Compatibility）**: 对于每一个对应维度，满足以下任一条件即可广播： `1`. 两个维度值相等 `2`. 其中一个维度为 1（会自动“扩展”）

```
a.shape = (3, 4, 5)
b.shape = (1, 4, 5) broadcast to (3, 4, 5)
```

3. **完成操作（Element-wise Operation）**：如果所有维度都满足条件，两个数组就被广播为相同形状，可以进行逐元素操作（如加法、乘法等）。

````python
#@title Exercises NumPy

arr_2d = np.arange(27).reshape(3, -1)
print(arr_2d)

# 任务 1：对每列进行中心化（减去列的均值）
col_means = arr_2d.mean(axis=0)                    # 沿 axis=0（即列）计算均值
centered_by_column = arr_2d - col_means            # 对每一列的每个元素减去该列的均值
print("\n任务1：\n", centered_by_column)


# 任务 2：将每一行归一化（总和为 1）
row_sums = arr_2d.sum(axis=1, keepdims=True)       # 计算每一行的总和，保持列结构（用于广播）
normalized_by_row = arr_2d / row_sums              # 每行除以它自己的和，实现归一化
print("\n任务2：\n", normalized_by_row)


# 任务 3：将每个元素替换为它与数组整体均值的距离
overall_mean = arr_2d.mean()                       # 计算整个数组的均值
distance_from_mean = np.abs(arr_2d - overall_mean) # 使用 np.abs 计算每个元素与整体均值的绝对距离
print("\n任务3：\n", distance_from_mean)
````
